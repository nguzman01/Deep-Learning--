Implementación de Redes Neuronales Básicas

```python

import numpy as np

def red_neuronal_basica(entradas, pesos, sesgo, umbral = 0.5):


  # calculo del puntaje z (suma ponderada + sesgo(bias))
  puntaje_z = np.dot(entradas, pesos) + sesgo

  # Redondeamos a 2 decimales evitando errores de precisión
  z_redondeado = round(puntaje_z, 2)

  # regla de activacion (Umbral en 0)
  # si el puntaje z es mayor al umbral, la salida es 1, si no es 0.

  salida = 1 if z_redondeado  >= umbral else 0

  return puntaje_z, salida

# conjunto de pruebas (casos controlados)
# se definen pesos y sesgos para probar el comportamiento

pesos = np.array([0.7, 0.3])
sesgo = -0.2

pruebas = [
    np.array([1, 1]),
    np.array([1, 0]),
    np.array([0, 1]),
    np.array([0, 0])
]

# Visualización de resultados
print(f"{'Prueba':<15} | {'Entradas':<15} | {'Puntaje z':<12} | {'Salida (0/1)':<12}")
print("-" * 65)

for i, entradas in enumerate(pruebas):
    z, clasificacion = red_neuronal_basica(entradas, pesos, sesgo)
    print(f"Caso {i+1:<10} | {str(entradas):<15} | {z:<12.2f} | {clasificacion:<12}")
```

## Conclusiones
A partir de la implementacion de la neurona basica, se pueden extraer las siguentes conclusiones sobre su comportamiento:

1. Efecto de los pesos:
los pesos definen la importancia de cada variable de entrada, en nuestra prueba al asignar un peso mayor a una de las emtradas, este tiene un impacto significativo a la hora de calcular el puntaje z.

2. Ahora, el sesgo  actua como un desplazador de la función de activación. un sesgo negativo como el que usamos de -0.2, aumentó la exigencia del modelo obligando que las suma de las entradas ponderadas sean mas altas logrando  asi la activación de la neurona.

3. Por ultimo, el umbral es el criterio final de la clasificación. aunque el puntaje z fuera positivo, si no supera el umbral en este caso el 0.5, la salida seria 0. Demostarndo asi como la red neuronal transforma datos continuos en una desición binaria (1/0). Es clave notar que si el puntaje z alcanza exactamente el valor del umbral (0.5), la neurona se activa, cumpliendo con la regla de decisión.
