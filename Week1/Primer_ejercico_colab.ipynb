```python
import numpy as up

def step(z):
  return 1 if z >= 0 else 0

def neurona(x1, x2, w1, w2 ,b):
  z = x1*w1 + x2*w2 + b
  y = step(z)
  return z, y



# Valores base = intenta parecerse a una compuerta ADN
w1, w2, b = 1, 1, -1.5

casos = [(0,0), (0,1), (1,0), (1,1)]
for x1, x2 in casos:
    z, y = neurona(x1, x2, w1, w2 , b)
    print( f"Entrada =({x1}, {x2}) z={z: .1f}  salida = {y}")

```

Preguntas:

*   Con que valor de b la neurona se comporta mas parecido a ADN?

    Rta: con la b = -2.0, pero tambien con la b= -1.5 se ve parecido a ADN,  porque algunas de sus entradas son 1 su salida da 1, y en este caso estas dos lo cumplen.

*   Que efecto tiene bajar o subir el bias en la salida?

    Rta: Entre mas bajo es el bias, es menos probable la activacion de la neurona y entre más arriba el bias, la neuroina se activa más facilmente.


En otras palabras cuando el bias es:

*   Poco negativo --> la neurona se activa más facilmente, por ejemplo  con el bias: b = -0.5
*   Más negativo --> la neurona necesita más suma para activarse, por ejemplo con el bias: b = -2.0
